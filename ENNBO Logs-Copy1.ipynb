```python
source='C:\\Users\\alcantara\\Documents\\ENNBO\\Logs\\MT4\\2018\\7'

#Palabra=['AUDUSD-PRO']
Palabra=['FIBONACHO V6.0 ESTABLE MT4 AUDUSD-PRO,M15: 0.','FIBONACHO V6.0 ESTABLE MT4 AUDUSD-PRO,M15: 1.','FIBONACHO V6.1 ESTABLE MT4 AUDUSD AUDUSD-PRO,M15: 0.','FIBONACHO V6.1 ESTABLE MT4 AUDUSD AUDUSD-PRO,M15: 1.']
#Palabra=['XAU 3.333 LSTM2 H4 PULVERIZAR2 XAUUSD-PRO,H4: 0.','XAU 3.333 LSTM2 H4 XAUUSD-PRO,H4: 0.','XAU 3.333 LSTM2 H4 PULVERIZAR2 XAUUSD-PRO,H4: 1.','XAU 3.333 LSTM2 H4 XAUUSD-PRO,H4: 1.']


for log in os.listdir(source):
    file=open(source+'\\'+log)
    lines=file.readlines()
    file.close()
    
    archivo=os.path.basename(source+'\\'+log)
    year=archivo[0:4]
    month=archivo[4:6]
    day=archivo[6:8]
    fecha=day+"/"+month+"/"+year
    reporte=open('C:\\Users\\alcantara\\Notebooks\\Logs ENNBO\\ReportesAUD\\ReporteAUDUSD.txt',"a")    
    #reporte=open('C:\\Users\\alcantara\\Notebooks\\Logs ENNBO\\Reportes\\ReporteXAUUSD.txt',"a")
        
    for line in lines:
        target=[]
        line=line.strip().upper()
        for p in Palabra:  
            valor=line.find(p)
            target.append(valor)
        target=np.max(target)
        if target>0:
            reporte.write(fecha)
            reporte.write("\t")
            reporte.write(line)
            reporte.write("\n")  
    reporte.close()
```

```python
source='C:\\Users\\alcantara\\Notebooks\\Logs ENNBO\\ReportesAUD'

Palabra=['FIBONACHO V6.0 ESTABLE MT4 AUDUSD-PRO,M15: 0.','FIBONACHO V6.0 ESTABLE MT4 AUDUSD-PRO,M15: 1.']


for log in os.listdir(source):
    file=open(source+'\\'+log)
    lines=file.readlines()
    file.close()
    
    archivo=os.path.basename(source+'\\'+log)
    #year=archivo[0:4]
    month=archivo[4:6]
    #day=archivo[6:8]
    #fecha=day+"/"+month+"/"+year
    
    #reporte=open('C:\\Users\\alcantara\\Notebooks\\Logs ENNBO\\Reportes\\ReporteJPYUSD.txt',"a")    
    reporte=open('ReporteAUD.txt',"a")
    filtro=0
    nofiltro=0
    
    for line in lines:
        target=[]
        line=line.strip().upper()
        for p in Palabra:  
            valor=line.find(p)
            target.append(valor)
        target=np.max(target)
        if target>0:
            
            e1=line[69:77] 
            if float(e1)<.5:
                filtro+=1
            if float(e1)>.5:
                nofiltro+=1

                
            #reporte.write(fecha)
            #reporte.write("\t")
            reporte.write(line)
            reporte.write("\n")
         
         
    reporte.write("Filtradas:")
    reporte.write(str(filtro))
    reporte.write("\n")  
    reporte.write("No Filtradas:")
    reporte.write(str(nofiltro))
    reporte.write("\n")
    reporte.close()
```
